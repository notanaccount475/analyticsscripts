```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Read Excel file into a DataFrame
df = pd.read_excel('C:/Users/StephanieLin/Downloads/benchmarks2.xlsx', sheet_name='Sheet1')  # Update the path and sheet name
benchmark_ids = df.iloc[:, 0].dropna().tolist()  # Extract the IDs
benchmarks = df.iloc[:, 1].dropna().tolist()  # Looks at the second column

# transform the text into vector form
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(benchmarks)


feature_names = tfidf.get_feature_names_out()
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)
tfidf_df
```

```python
# cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
similarity_df = pd.DataFrame(cosine_sim, index=benchmarks, columns=benchmarks)


similarity_df.head()
```

```python
# Loop through the matrix to find similar rows
threshold = 0.5  # set a threshold for similarity
similar_pairs = []

for i in range(len(cosine_sim)):
    for j in range(i + 1, len(cosine_sim)):
        if cosine_sim[i][j] > threshold:
            similar_pairs.append({
                "ID1": benchmark_ids[i],
                "Benchmark1": benchmarks[i],
                "ID2": benchmark_ids[j],
                "Benchmark2": benchmarks[j]
            })

similar_pairs_df = pd.DataFrame(similar_pairs)
similar_pairs_df
```

```python
similar_pairs_df.to_excel('similar_pairs.xlsx', index=False)
```
